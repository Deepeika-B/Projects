

**Login system using file handling**

```python

import re

def register():
   uname=input("Enter the username:")
   pw=input("Enter the password:")
   pat = "^[a-zA-Z0-9-_]+@[a-zA-Z0-9]+\.[a-z]{1,3}$"
   pwpat= "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?& ])[A-Za-z\d@$!#%*?&]{5,16}$"
   if re.match(pat,uname) and re.match(pwpat,pw):
     fname=open("data.txt","a+")
     fname.write(uname+", "+pw+"\n")
     fname.close()
     print("Valid username and password")
     print("Registration successful")
   elif re.match(pat,uname):
     print("Invalid password")
   elif re.match(pwpat,pw):
     print("Invalid username")
   else:
     print("Invalid username and password")
   
def forgotPassword():
  fname=open("data.txt","r")
  uname=input("Enter the username:")
  u=[]
  p=[]
  for each in fname:
    a,b=each.split(",")
    b=b.strip()
    u.append(a)
    p.append(b)
  fname.close()
  data1=dict(zip(u,p))
  if uname in u:
    print("Your password is:", data1[uname])
  else:
    print("Username doesn't exist...Register to continue")
    register()

def login():
  uname=input("Enter the username:")
  pw=input("Enter the password:")
  fname=open("data.txt","r")
  u=[]
  p=[]
  for each in fname:
    a,b=each.split(",")
    b=b.strip()
    u.append(a)
    p.append(b)
  fname.close()
  data1=dict(zip(u,p))
  if uname in u:
    if pw == data1[uname]:
      print("Login successful")
    else:
      print("Incorrect password")
      forgotPassword()
  else:
    print("Invalid username")
    register()

a=input("Choose to register or to login")
if a== "R":
  register()
elif a == "L":
  login()


    


```
